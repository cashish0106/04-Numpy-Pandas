```python
import pandas as pd
import numpy as np
```

```python
purchase_df=pd.read_json("purchase_data.json")
purchase_df.head()
```

```python
#Player Count
Total_Players=purchase_df["SN"].nunique()
stat_1=pd.DataFrame({"Total Players":[Total_Players]})
stat_1
```

```python
#Gender Demographics
Total_Revenue=purchase_df["Price"].sum()
Avg_Price=purchase_df["Price"].mean()
Num_Purchase=purchase_df["Price"].count()
Unique_Item=len(purchase_df["Item ID"].unique())

stat_1["Total Revenue"]=Total_Revenue
stat_1["Average Price"]=Avg_Price
stat_1["Number of Purchases"]=Num_Purchase
stat_1["Number of Unique Items"]=Unique_Item
stat_1[["Average Price","Total Revenue"]]=stat_1[["Average Price","Total Revenue"]].applymap("${:,.2f}".format)
stat_1[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
```

```python
#Gender Demographics
gender_data_df=purchase_df[["SN","Gender"]]
gender_data_df.drop_duplicates(inplace=True)
stat_2=pd.DataFrame(gender_data_df.groupby("Gender").count())
stat_2.rename(columns={"SN":"Total Count"},inplace=True)
stat_2["Percentage of Players"]=(stat_2["Total Count"]/gender_data_df["SN"].count()*100).map("{:.2f}".format)
stat_2[["Percentage of Players","Total Count"]]
```

```python
#Purchasing Analysis (Gender)
purchase_df.head()
gender_group=purchase_df.groupby("Gender")
avg_price=gender_group["Price"].mean()
sum_price=gender_group["Price"].sum()
Norm_total=sum_price/stat_2["Total Count"]
stat_3=pd.DataFrame(purchase_df["Gender"].value_counts())
stat_3.rename(columns={"Gender":"Purchase Count"},inplace=True)
stat_3["Average Purchase Price"]=avg_price
stat_3["Total Purchase Value"]=sum_price
stat_3["Normalized Totals"]=Norm_total
stat_3[["Average Purchase Price","Total Purchase Value","Normalized Totals"]]=stat_3[["Average Purchase Price","Total Purchase Value","Normalized Totals"]].applymap("${:,.2f}".format)
stat_3
```

```python
#Age Demographics
age_data_df=purchase_df[["SN","Age"]]
age_data_df.drop_duplicates(inplace=True)
bins=[0,9,14,19,24,29,34,39,100]
group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
age_data_df["Age Demographics"]=pd.cut(age_data_df["Age"],bins,labels=group_labels)
stat_4=pd.DataFrame(age_data_df.groupby("Age Demographics").count())
stat_4.drop("SN",inplace=True,axis=1)
stat_4.rename(columns={"Age":"Total Count"},inplace=True)
stat_4["Percentage of Players"]=(stat_4["Total Count"]/age_data_df["Age"].count()*100).map("{:.2f}".format)
stat_4
```

```python
#Purchasing Analysis (Age)
purchase_data_df=purchase_df[["Age","Price"]]
bins=[0,9,14,19,24,29,34,39,100]
group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data_df["Age Demographics"]=pd.cut(purchase_data_df["Age"],bins,labels=group_labels)
stat_5=pd.DataFrame(purchase_data_df.groupby("Age Demographics").count())
stat_5.drop("Price",inplace=True,axis=1)
stat_5.rename(columns={"Age":"Purchase Count"},inplace=True)
stat_5.head()
stat_5["Average Purchase Price"]=(purchase_data_df.groupby("Age Demographics")["Price"].mean()).map("${:,.2f}".format)
stat_5["Total Purchase Value"]=(purchase_data_df.groupby("Age Demographics")["Price"].sum())
stat_5["Normalized Totals"]=(stat_5["Total Purchase Value"]/stat_4["Total Count"]).map("${:,.2f}".format)
stat_5["Total Purchase Value"]=stat_5["Total Purchase Value"].map("${:,.2f}".format)
stat_5
```

```python
#Top Spenders
purchase_df.head()
stat_6=pd.DataFrame(purchase_df["SN"].value_counts())
stat_6.rename(columns={"SN":"Purchase Count"},inplace=True)
stat_6["Average Purchase Price"]=(purchase_df.groupby("SN")["Price"].mean()).map("${:,.2f}".format)
stat_6["Total Purchase Value"]=(purchase_df.groupby("SN")["Price"].sum()).map("${:,.2f}".format)
stat_6.head()
```

```python
#Most Popular Items

#1. Finding unique ITEM NAMe and Item ID
item_df=purchase_df[["Item Name","Item ID","Price"]]
item_df.drop_duplicates(inplace=True)
item_df.count()

stat_7=pd.DataFrame(purchase_df["Item ID"].value_counts())
stat_7.rename(columns={"Item ID":"Purchase Count"},inplace=True)
stat_7["Total Purchase Value"]=(purchase_df.groupby("Item ID")["Price"].sum())
stat_7.reset_index(inplace=True)
stat_7=stat_7.merge(item_df,left_on="index",right_on="Item ID")
pd.options.display.float_format = '${:,.2f}'.format
stat_7.sort_values("Purchase Count",ascending=False)[["Item ID","Item Name","Purchase Count","Price","Total Purchase Value"]].head(5)
#.map("${:,.2f}".format)
```
